# Historical DWD data analysis

```python
import pandas as pd
import geopandas as gpd
import numpy as np
from pathlib import Path
from agrotools import configs
from shapely.geometry import Point
import matplotlib.pyplot as plt
from pylab import rcParams
from mtranslate import translate
from tqdm import tqdm

%matplotlib inline
```

```python
prjconf = configs.ProjectConfigParser(config_file=None)

gis_raw = prjconf.get_path('GISdata', 'raw')
gis_dwd = prjconf.get_path('GISdata', 'dwd')
```

#### functions

```python
def get_clean_df(csv_path):
    df = pd.read_csv(csv_path, sep=';')
    df= df.iloc[:-1,:-2]
    df.columns = df.columns.str.strip()
    df['Eintrittsdatum'] = pd.to_datetime(df.Eintrittsdatum, format='%Y%m%d')
    assert len(df.Objekt_id.unique()) == 1
    object_id = df.Objekt_id.unique()[0]
    print(df.shape)
    display(df.tail(1))

    dfs = []
    for (stationid, phase_id), _df in tqdm(df.groupby(['Stations_id', 'Phase_id'], group_keys=False)):
        _med = _df.Jultag.median()
        numb_years = len(_df.Referenzjahr.unique())
        dfs.append(pd.DataFrame([int(stationid), int(phase_id), int(_med), numb_years]).T.rename(columns={0:'Stations_id', 1:'Phase_id', 2:'Jultag_med', 3:'Referenzjahr_num'}))
    df_med = pd.concat(dfs)
    display(df_med.head(1))
    print('After calculating medians', df_med.shape)

    df_phases = df_crop_phases[df_crop_phases.Objekt_id == object_id][['Phasen_id', 'Phase', 'Phase_en', 'Phasendefinition', 'Phasendefinition_en']]
    df_phases['Phasen_id'] = df_phases.Phasen_id.astype('int64')
    display(df_phases)
    
    df_med = df_med.merge(df_phases, how='left', left_on='Phase_id', right_on='Phasen_id')
    print(df_med.shape)
    df_med = df_med.merge(df_stations[['Stations_i', 'Stationsna', 'Bundesland', 'geometry']], how='inner', left_on='Stations_id', right_on='Stations_i')
    df_med = gpd.GeoDataFrame(df_med, geometry=df_med.geometry, crs='EPSG:32632')
    print(df_med.shape)
    display(df_med.head(1))
    return df_med
```

## Maize

```python
df_maize = pd.read_csv(Path(gis_dwd/'phenology'/'PH_Sofortmelder_Landwirtschaft_Kulturpflanze_Mais_ohne_Sortenangabe_1979_2018_hist.txt'), sep=';')
df_maize = df_maize.iloc[:-1,:-2]
df_maize.columns = df_maize.columns.str.strip()
df_maize['Eintrittsdatum'] = pd.to_datetime(df_maize.Eintrittsdatum, format='%Y%m%d')
print(df_maize.shape)
display(df_maize.head(2))
display(df_maize.tail(2))
```

```python
dfs = []
for (stationid, phase_id), _df in tqdm(df_maize.groupby(['Stations_id', 'Phase_id'], group_keys=False)):
    _med = _df.Jultag.median()
    numb_years = len(_df.Referenzjahr.unique())
    dfs.append(pd.DataFrame([int(stationid), int(phase_id), int(_med), numb_years]).T.rename(columns={0:'Stations_id', 1:'Phase_id', 2:'Jultag_med', 3:'Referenzjahr_num'}))
df_maize_med = pd.concat(dfs)
display(df_maize_med.head(3))
print(df_maize_med.shape)
```

```python
df_phases = df_crop_phases[df_crop_phases.Objekt_id == 210][['Phasen_id', 'Phase', 'Phase_en', 'Phasendefinition', 'Phasendefinition_en']]
df_phases['Phasen_id'] = df_phases.Phasen_id.astype('int64')
display(df_phases.head(2))
print(df_phases.shape)
```

```python
df_maize_med = df_maize_med.merge(df_phases, how='left', left_on='Phase_id', right_on='Phasen_id')
print(df_maize_med.shape)
df_maize_med = df_maize_med.merge(df_stations[['Stations_i', 'Stationsna', 'Bundesland', 'geometry']], how='inner', left_on='Stations_id', right_on='Stations_i')
print(df_maize_med.shape)
display(df_maize_med.head(3))
```

```python
gpd.GeoDataFrame(df_maize_med, geometry=df_maize_med.geometry, crs='EPSG:32632').to_file(Path(gis_dwd/'phenology'/'shp'/'PH__hist_1979_2017__maize.shp'))
```
